# resize 變數

```python
x_train_resize = []
```

```python
for index,uid in enumerate(train_metadata['UID']):
    print('處理 UID:',uid,',進度',index+1,'/',train_metadata['UID'].shape[0])
    ecg_tmp = np.array(train_data.iloc[index,:]).reshape(1000,12)
    ecg_tmp = np.repeat(ecg_tmp,20,axis=1)
    x_train_resize.append(ecg_tmp)
```

```python
x_train_resize = np.array(x_train_resize)
```

```python
x_train_resize.shape
```

```python
np.save('dataset/x_train_resize.npy', x_train_resize)
x_train_resize = np.load('dataset/x_train_resize.npy')
```
